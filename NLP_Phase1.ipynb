# 🤔 TOPIC 1.1 — What is NLP?

## Simple Definition
**NLP = Teaching computers to read, understand, and write human language.**

## Real Life Examples You Already Use
| When you... | NLP is working behind the scenes |
|-------------|----------------------------------|
| Type in Google Search | NLP understands your query |
| Get spam filtered in Gmail | NLP detects spam |
| Use Google Translate | NLP translates language |
| Talk to Siri / Alexa | NLP converts speech to text + understands it |
| Chat with a customer bot | NLP powers the chatbot |
| See 'Recommended for you' | NLP read your past activity |

## Why is it Hard for Computers?
Computers love numbers. Text is messy. Language is full of:
- **Ambiguity** (same word, different meaning)
- **Sarcasm** ("Oh great, another Monday 🙄")
- **Slang** ("That movie was fire 🔥")
- **Typos** ("I luv crikt")
- **Context** ("I went to the bank" — river bank or money bank?)

Let's see ambiguity in code:

````python
# Let's show why NLP is hard — the same word means different things!

# We make a list of sentences. A list uses square brackets []
# Each item in the list is a sentence (text inside quotes)
ambiguous_examples = [
    "I went to the bank.",               # bank = river bank OR money bank?
    "The bark was very loud.",            # bark = dog sound OR tree bark?
    "She can't bear the pain.",           # bear = animal OR tolerate?
    "I saw the man with the telescope.",  # who has the telescope — you or the man?
    "Flying planes can be dangerous.",   # are you flying them, or are the planes dangerous?
]

# Now let's print each example with a number
# enumerate() gives us both the index (number) and the item from the list
# start=1 means start counting from 1 (not 0)
print("=" * 55)
print("  WHY NLP IS HARD — AMBIGUOUS SENTENCES")
print("=" * 55)

# 'for' loop goes through each item in the list, one by one
# i = the number (1, 2, 3...)
# sentence = the actual text
for i, sentence in enumerate(ambiguous_examples, start=1):
    # f-string: f"..." lets you put variables inside curly braces {}
    print(f"\nExample {i}: {sentence}")
    print(f"           ❓ What does it actually mean?")

print("\n" + "=" * 55)
print("Humans understand these from context.")
print("Teaching MACHINES to do the same = NLP!")
````

# 🧹 TOPIC 1.2 — Text Cleaning

## What is Text Cleaning?
Real-world text is MESSY. Before any NLP, we must clean it.

**Common garbage in text:**
- HTML tags: `<p>`, `<b>`, `<br>`
- URLs: `https://www.youtube.com`
- Hashtags: `#cricket #india`
- Extra spaces: `hello    world`
- Special characters: `!!!`, `@@@`, `***`
- Numbers (sometimes)

## Tool We Use: Regular Expressions (`re`)
Regular expressions (regex) are patterns used to find and remove text.
- `re.sub(pattern, replacement, text)` = find the pattern in text, replace with replacement
- Example: `re.sub(r'<.*?>', '', text)` → removes all HTML tags

````python
# First, let's create a messy, real-world looking text
# This is the kind of text you get from websites, social media, etc.

messy_text = """<p>OMG!!! Check out this AMAZING cricket match 🏏🔥
Visit https://cricbuzz.com for live scores!!!
India beat Pakistan by 100 runs   today in Mumbai.
@Virat_Kohli scored a century... #Cricket #India #WorldCup2023</p>"""

# Let's see what it looks like before cleaning
print("BEFORE CLEANING:")
print("-" * 50)
print(messy_text)
print()
````

````python
# Now let's clean the text — ONE STEP AT A TIME
# We will print after each step so you can see exactly what changed

# We start with the original messy text
text = messy_text

# ------ STEP 1: Remove HTML tags ------
# HTML tags look like: <p>, <b>, <div>, <br>
# Pattern r'<.*?>' means: find anything that starts with < and ends with >
# r'' = raw string (just means backslashes are treated literally)
# '.*?' = any characters (.) zero or more times (*) but stop at first > (? = lazy match)
# We replace the found pattern with '' (empty string = delete it)
text = re.sub(r'<.*?>', '', text)
print("After removing HTML tags:")
print(text)
print()
````

````python
# ------ STEP 2: Remove URLs ------
# URLs start with http or www
# Pattern: r'https?://\S+' means:
#   http  = the literal text 'http'
#   s?    = the letter s is optional (covers both http and https)
#   ://   = literal characters ://
#   \S+   = one or more non-space characters (the rest of the URL)
# r'www\.\S+' catches URLs starting with www
text = re.sub(r'https?://\S+|www\.\S+', '', text)
print("After removing URLs:")
print(text)
print()
````

````python
# ------ STEP 3: Remove @mentions ------
# Pattern: r'@\w+' means:
#   @   = the @ symbol
#   \w+ = one or more word characters (letters, digits, underscore)
# This catches things like @Virat_Kohli, @username
text = re.sub(r'@\w+', '', text)
print("After removing @mentions:")
print(text)
print()
````

````python
# ------ STEP 4: Remove #hashtags ------
# Pattern: r'#\w+' means:
#   #   = the hash symbol
#   \w+ = one or more word characters after it
# This catches: #Cricket, #India, #WorldCup2023
text = re.sub(r'#\w+', '', text)
print("After removing hashtags:")
print(text)
print()
````

````python
# ------ STEP 5: Remove special characters ------
# We want to KEEP only: letters (a-z, A-Z), digits (0-9), spaces
# Pattern: r'[^a-zA-Z0-9\s]' means:
#   [^...] = NOT any of these characters
#   a-zA-Z = any English letter
#   0-9    = any digit
#   \s     = any space/tab/newline
# So we remove everything that is NOT a letter, digit, or space
text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
print("After removing special characters (!!! 🏏 🔥 ...):")
print(text)
print()
````

````python
# ------ STEP 6: Remove extra whitespace ------
# After removing things, we often get multiple spaces or newlines
# Pattern: r'\s+' means: one or more whitespace characters (space, tab, newline)
# We replace all that with a single space ' '
text = re.sub(r'\s+', ' ', text)

# .strip() removes any space at the very beginning or very end of the text
text = text.strip()

print("After removing extra spaces:")
print(text)
print()

print("=" * 50)
print("SUMMARY:")
print(f"BEFORE: {messy_text[:80]}...")
print(f"AFTER:  {text}")
````

````python
# Now let's wrap all 6 steps into ONE reusable function
# A function is a block of code you can call again and again with different input

# 'def' means we are defining (creating) a function
# 'clean_text' is the name we give to this function
# 'text' inside the brackets is the INPUT — whatever text we pass in
def clean_text(text):
    # Step 1: Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Step 2: Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    # Step 3: Remove @mentions
    text = re.sub(r'@\w+', '', text)
    # Step 4: Remove #hashtags
    text = re.sub(r'#\w+', '', text)
    # Step 5: Remove special characters (keep letters, digits, spaces)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Step 6: Remove extra whitespace and trim edges
    text = re.sub(r'\s+', ' ', text).strip()
    # 'return' sends back the cleaned text to whoever called this function
    return text

# Let's test our function on a new example
test_sentence = "<h1>Hello World!!!</h1> Visit https://example.com @user #tag 🎉"

print("Testing our clean_text() function:")
print(f"Input:  {test_sentence}")
print(f"Output: {clean_text(test_sentence)}")
print("\n✅ Our function works! We can now clean any text with one line.")
````

# ✂️ TOPIC 1.3 — Tokenization

## What is Tokenization?
**Tokenization = Breaking text into smaller pieces called tokens.**

A token is usually a word, but it can also be a sentence, a character, or a sub-word.

**Why do we need it?**
Computers cannot process a whole paragraph at once. We need to break it into pieces first.

**Two types:**
1. **Word Tokenization** → Split into words: `"I love NLP"` → `["I", "love", "NLP"]`
2. **Sentence Tokenization** → Split into sentences: `"I love NLP. It is fun."` → `["I love NLP.", "It is fun."]`

````python
# Let's use this sample paragraph for tokenization
sample_paragraph = """India is a beautiful country.
It has a population of 1.4 billion people.
India won the Cricket World Cup in 1983 and 2011."""

print("Our paragraph:")
print(sample_paragraph)
print()

# ---- SENTENCE TOKENIZATION ----
# sent_tokenize() splits text into a LIST of sentences
# It uses punctuation (. ! ?) to decide where sentences end
sentences = sent_tokenize(sample_paragraph)

print("=" * 50)
print("SENTENCE TOKENIZATION:")
print("=" * 50)
# len() gives us the number of items in a list
print(f"Number of sentences found: {len(sentences)}")
print()

# Loop through each sentence and print it
for i, sentence in enumerate(sentences, start=1):
    print(f"Sentence {i}: {sentence}")
print()
````

````python
# ---- WORD TOKENIZATION ----
# word_tokenize() splits text into a LIST of individual words (and punctuation)
# It is smarter than just splitting by spaces
# For example: "don't" → ["do", "n't"] (it knows about contractions)

words = word_tokenize(sample_paragraph)

print("=" * 50)
print("WORD TOKENIZATION:")
print("=" * 50)
print(f"Number of tokens found: {len(words)}")
print()
print("All tokens:")
print(words)   # This prints the full list with brackets
print()

# Let's also show them one by one, nicely
print("Tokens one by one:")
for i, word in enumerate(words, start=1):
    # end=' ' means print them on the same line with a space between
    print(f"{word}", end=' | ')

print("\n")
print("Notice: Punctuation (. ,) is also treated as a separate token!")
````

````python
# IMPORTANT: word_tokenize vs split()
# Beginners often use .split() to split words.
# Let's see why word_tokenize is BETTER.

test = "I don't like this movie, it's boring!"

# .split() just splits on spaces — it treats "don't" as ONE token
split_result = test.split()

# word_tokenize is smarter — it splits "don't" into ["do", "n't"]
tokenize_result = word_tokenize(test)

print("Comparing split() vs word_tokenize()")
print(f"\nText: '{test}'")
print(f"\nUsing .split()        → {split_result}")
print(f"\nUsing word_tokenize() → {tokenize_result}")
print()
print("✅ word_tokenize correctly handles contractions, punctuation, etc.")
print("   Always use word_tokenize() in NLP projects!")
````

# 🗑️ TOPIC 1.4 — Stopword Removal

## What are Stopwords?
Stopwords are **very common words** that appear everywhere but carry **no real meaning**.

Examples: `the`, `is`, `are`, `a`, `an`, `in`, `on`, `I`, `we`, `this`, `that`...

## Why Remove Them?
- They add noise without adding meaning
- They make our vocabulary bigger (waste of memory)
- They confuse models

**Example:**
- Before: `"The movie is very good and I loved it"`
- After: `["movie", "good", "loved"]` ← These are the meaningful words!

> Note: In some tasks (like translation), we do NOT remove stopwords. Remove them only when you need the key meaning.

````python
# First, let's see what stopwords look like
# stopwords.words('english') returns a LIST of all English stopwords
english_stopwords = stopwords.words('english')

print("Total stopwords in NLTK's English list:", len(english_stopwords))
print()

# sorted() arranges the list alphabetically
# [:30] means: give me only the first 30 items from the list
print("First 30 stopwords (sorted alphabetically):")
print(sorted(english_stopwords)[:30])
````

````python
# Now let's remove stopwords from a real sentence

sentence = "The movie was very good and I really loved watching it with my family"

# Step 1: Tokenize the sentence into words
tokens = word_tokenize(sentence)
print("Original tokens:")
print(tokens)
print()

# Step 2: Convert stopwords list to a SET
# set() is faster than list for checking "is this word in the list?"
# We use set() because we check thousands of words — sets are much faster
stop_words_set = set(english_stopwords)

# Step 3: Keep only words that are NOT stopwords
# Also convert to lowercase first (.lower()) so 'The' and 'the' are treated the same
# 'if word.lower() not in stop_words_set' = only keep if it's NOT a stopword
filtered_tokens = [word for word in tokens if word.lower() not in stop_words_set]
# The above is called a LIST COMPREHENSION — a short way to filter a list
# It is the same as writing:
#   filtered_tokens = []
#   for word in tokens:
#       if word.lower() not in stop_words_set:
#           filtered_tokens.append(word)

print("After removing stopwords:")
print(filtered_tokens)
print()
print(f"Went from {len(tokens)} tokens → {len(filtered_tokens)} tokens")
print(f"Removed {len(tokens) - len(filtered_tokens)} stopwords!")
````

````python
# BONUS: You can add YOUR OWN custom stopwords!
# For example, if you are analyzing cricket news, you might want to remove 'cricket'

# Start with the standard NLTK stopwords
custom_stopwords = set(stopwords.words('english'))

# .update() adds new words to the set
# We add words that are very common in cricket news but not meaningful for our analysis
custom_stopwords.update(['cricket', 'match', 'player', 'team', 'game'])

# Test sentence about cricket
cricket_text = "India cricket team won the match in a thrilling game yesterday"
cricket_tokens = word_tokenize(cricket_text)

# Remove both standard + custom stopwords
filtered = [w for w in cricket_tokens if w.lower() not in custom_stopwords]

print("Custom Stopword Removal Example:")
print(f"Original: {cricket_tokens}")
print(f"Filtered: {filtered}")
print()
print("💡 Tip: Always customize stopwords based on your specific task!")
````

# ✂️ TOPIC 1.5 — Stemming

## What is Stemming?
Stemming **cuts off the ending of a word** to get its base/root form.

**Examples:**
- `running` → `run`
- `played` → `play`
- `studies` → `studi` ← ⚠️ Not always a real word!

## Why Do We Need It?
To a computer, `run`, `running`, `runs`, `ran` are FOUR DIFFERENT words.
But they all mean the same thing! Stemming maps them all to `run`.

## Limitation
Stemming is **fast but crude**. It doesn't always give a real word.
- `better` → `better` (can't stem this correctly)
- `studies` → `studi` (not a real word!)

````python
# Create a PorterStemmer object
# PorterStemmer is the most popular stemming algorithm, created by Martin Porter in 1980
stemmer = PorterStemmer()

# A list of words to test stemming on
# These are all different forms of the same base word
words_to_stem = [
    # Verb forms
    'run', 'running', 'runs', 'ran',
    # More verb forms
    'play', 'playing', 'played', 'plays',
    # Adjective forms
    'happy', 'happiness', 'happily',
    # Tricky ones
    'studies', 'studying', 'studied',
    'better', 'best', 'good',
]

print("=" * 45)
print("       STEMMING RESULTS")
print("=" * 45)
# Print a header row using f-string formatting
# :<15 means: left-align and take at least 15 characters (for neat columns)
print(f"{'Original Word':<20} → {'Stemmed Form'}")
print("-" * 45)

# stemmer.stem(word) returns the stemmed version of that word
for word in words_to_stem:
    stemmed = stemmer.stem(word)  # This does the actual stemming
    print(f"{word:<20} → {stemmed}")

print()
print("⚠️  Notice: 'studies' → 'studi' is NOT a real word!")
print("   This is the main problem with stemming.")
print("   Solution: Use Lemmatization (next topic)")
````

# 🌿 TOPIC 1.6 — Lemmatization

## What is Lemmatization?
Lemmatization also reduces words to their base form, but it does it **properly**.
It uses a **dictionary** to find the real base form (called the **lemma**).

**Examples:**
- `running` → `run` ✅ (real word)
- `studies` → `study` ✅ (real word, not 'studi'!)
- `better` → `good` ✅ (knows 'better' is comparative form of 'good')

## Stemming vs Lemmatization
| | Stemming | Lemmatization |
|--|----------|---------------|
| **Speed** | Fast | Slower |
| **Result** | May not be a real word | Always a real word |
| **Uses dictionary** | No | Yes |
| **Example** | studies → studi | studies → study |
| **Use when** | Speed matters | Accuracy matters |

## Important: POS (Part of Speech) matters!
You should tell the lemmatizer what type of word it is:
- `pos='v'` → treat as a VERB (`running` → `run`)
- `pos='n'` → treat as a NOUN (`dogs` → `dog`)
- `pos='a'` → treat as an ADJECTIVE (`better` → `good`)

````python
# Create a WordNetLemmatizer object
# WordNet is a large database of English words and their relationships
lemmatizer = WordNetLemmatizer()

print("=" * 60)
print("   STEMMING vs LEMMATIZATION — Side by Side Comparison")
print("=" * 60)
print(f"{'Word':<15} {'Stemmed':<15} {'Lemmatized (verb)'}")
print("-" * 50)

# List of words to compare
test_words = ['running', 'played', 'studies', 'better', 'wolves', 'loving', 'geese', 'feet']

for word in test_words:
    stemmed = stemmer.stem(word)               # Stemming result
    lemmatized = lemmatizer.lemmatize(word, pos='v')  # Lemmatization result (as verb)
    # pos='v' tells the lemmatizer to treat this word as a verb
    print(f"{word:<15} {stemmed:<15} {lemmatized}")

print()
print("✅ Lemmatization always gives a REAL dictionary word!")
````

````python
# Understanding how POS (Part of Speech) affects lemmatization

word = 'better'

print("How POS tag changes lemmatization result:")
print(f"Word: '{word}'")
print()

# Without POS — lemmatizer doesn't know what type of word it is
# Default is 'n' (noun)
result_no_pos = lemmatizer.lemmatize(word)
print(f"No POS specified (defaults to noun): '{result_no_pos}'")

# pos='v' = verb
result_verb = lemmatizer.lemmatize(word, pos='v')
print(f"As a verb (pos='v')               : '{result_verb}'")

# pos='a' = adjective
result_adj = lemmatizer.lemmatize(word, pos='a')
print(f"As an adjective (pos='a')         : '{result_adj}'   ← 'good' is correct!")

print()
print("💡 Moral: Always specify the POS for best results.")
print("   In Phase 2, we will learn to auto-detect POS using POS Tagging!")
````

````python
# Lemmatization works great for nouns too!
# pos='n' means treat the word as a noun

nouns = ['dogs', 'wolves', 'children', 'geese', 'feet', 'teeth', 'mice', 'cacti']

print("Lemmatization for Nouns (plural → singular):")
print(f"{'Plural':<15} → {'Singular (lemma)'}")
print("-" * 35)

for noun in nouns:
    # pos='n' = noun
    lemma = lemmatizer.lemmatize(noun, pos='n')
    print(f"{noun:<15} → {lemma}")
````

# 🔁 TOPIC 1.7 — The Complete NLP Pipeline

## Putting Everything Together!

Now we will chain all the steps we learned into **one complete pipeline**.

```
Raw Text
   ↓
1. Clean Text        (remove HTML, URLs, etc.)
   ↓
2. Tokenize          (split into words)
   ↓
3. Lowercase         (INDIA → india)
   ↓
4. Remove Stopwords  (remove 'the', 'is', 'a'...)
   ↓
5. Remove Punctuation (remove '.', ',', '!'...)
   ↓
6. Lemmatize         (running → run, studies → study)
   ↓
Clean, Processed Text ✅
```

````python
# Let's build the full pipeline as a function
# We will print each step's output so you can follow along

def nlp_pipeline(raw_text, show_steps=True):
    """
    Complete NLP Preprocessing Pipeline.

    Parameters:
        raw_text (str)   : The input text to process
        show_steps (bool): If True, prints output after each step

    Returns:
        list: Final list of clean, processed tokens
    """

    if show_steps:
        print("INPUT TEXT:")
        print(f"  {raw_text}")
        print()

    # ---- STEP 1: Clean the text ----
    # Remove HTML, URLs, mentions, hashtags, special characters
    cleaned = re.sub(r'<.*?>', '', raw_text)           # Remove HTML tags
    cleaned = re.sub(r'https?://\S+|www\.\S+', '', cleaned)  # Remove URLs
    cleaned = re.sub(r'@\w+|#\w+', '', cleaned)       # Remove @mentions and #hashtags
    cleaned = re.sub(r'[^a-zA-Z0-9\s]', '', cleaned)  # Remove special characters
    cleaned = re.sub(r'\s+', ' ', cleaned).strip()    # Remove extra spaces

    if show_steps:
        print("STEP 1 — After Cleaning:")
        print(f"  {cleaned}")
        print()

    # ---- STEP 2: Tokenize ----
    # Split the cleaned text into individual word tokens
    tokens = word_tokenize(cleaned)

    if show_steps:
        print(f"STEP 2 — After Tokenization ({len(tokens)} tokens):")
        print(f"  {tokens}")
        print()

    # ---- STEP 3: Lowercase ----
    # Convert every token to lowercase
    # This ensures 'India' and 'india' are treated as the same word
    tokens = [token.lower() for token in tokens]

    if show_steps:
        print("STEP 3 — After Lowercasing:")
        print(f"  {tokens}")
        print()

    # ---- STEP 4: Remove Stopwords + Punctuation ----
    stop_words = set(stopwords.words('english'))  # Load the stopwords set
    punct = set(string.punctuation)               # Load all punctuation: .,!?;:'"...

    # Keep a token only if:
    #   - it is NOT a stopword
    #   - it is NOT punctuation
    #   - it contains only alphabetical characters (no numbers)
    tokens = [
        token for token in tokens
        if token not in stop_words      # not a stopword
        and token not in punct          # not punctuation
        and token.isalpha()             # only alphabetic (skip '100', '2023', etc.)
    ]

    if show_steps:
        print(f"STEP 4 — After Stopword & Punctuation Removal ({len(tokens)} tokens):")
        print(f"  {tokens}")
        print()

    # ---- STEP 5: Lemmatize ----
    lemmatizer = WordNetLemmatizer()  # Create lemmatizer object

    # Lemmatize each token as a verb (pos='v')
    # For a more accurate pipeline, you would first POS tag, then lemmatize
    # We will do that in Phase 2!
    tokens = [lemmatizer.lemmatize(token, pos='v') for token in tokens]

    if show_steps:
        print(f"STEP 5 — After Lemmatization ({len(tokens)} tokens):")
        print(f"  {tokens}")
        print()
        print("=" * 55)
        print("✅ FINAL PROCESSED TOKENS:")
        print(f"   {tokens}")
        print("=" * 55)

    # Return the final list of tokens
    return tokens


print("✅ nlp_pipeline() function is ready!")
print("Run the next cell to test it.")
````
